### Indexing 
views of a numpy array

```python
a = numpy.arange(-10,10)
a
```

```python
#b = a[3:10]      # elements 3 to 10
#b = a[[1,3,5,7]] # elements 1,3,5,7
#b = a[:5]        # first 5 elements
#b = a[0]         # first element
#b = a[-1]        # last element
#b = a[-10:]      # last 10 elements
#b = a[::-1]      # reverse order
#b = a[::2]       # every second element
#b = a[1::2]      # every second element starting from the index 1

print(b)
```

```python
# Views Vs. copies
# http://scipy-cookbook.readthedocs.io/items/ViewsVsCopies.html
# It's important to know that in the above commands b is not a new array, but a 'view' of a.
# i.e changing a will change b and vice versa
print('View:')
a = numpy.arange(-10,10)
b = a[-10:]
print(b)
a[-10:] = 0
print(b, '\n')

# To create a copy use numpy.copy():
print('Copy:')
a = numpy.arange(-10,10)
b = a[-10:].copy()
print(b)
a[-10:] = 0
print(b, '\n')

# One exception is b = a[[1,3,5,7]], which creates a  copy by default. 
# (it's considered 'fancy indexing', or not a slice, see link above)
print('fancy indexing:')
a = numpy.arange(-10,10)
b = a[[1,3,5,7]]
print(b)
a[[1,3,5,7]] = 0
print(b)
```

```python
# Also note that to free the memory of a variable one needs to delete all views
print('View:')
a = numpy.arange(-10,10)
b = a[-10:]
print(b)
del(a)
print(b, '\n')
```

```python
#Draw numbers from a normal distributions
#loc = mean
#scale = sigma
n1 = numpy.random.normal(loc = 0, scale=  1, size=10000)
n2 = numpy.random.normal(loc = 0, scale=  2, size=10000)
```

```python
plt.figure()
plt.hist(n1, bins = 100, normed=True, alpha = 0.5, color = 'blue',   label = r'$\sigma = 1$')
plt.hist(n2, bins = 100, normed=True, alpha = 0.5, color = 'purple', label = r'$\sigma = 2$')
plt.legend()
plt.show()
```

```python
# Sub-plots
# Example for a grid of plots in one figure, the size of the grid is controlled by nrows, ncols
# sharex - same x axis for all subplots
f, axes = plt.subplots(ncols=3, nrows=3, figsize = (10,7), sharex=True)

sample_size = 100
for ax in axes.flatten():
    n = numpy.random.normal(loc = 0, scale=  1, size=sample_size)
    
    ax.hist(n, bins = 100, normed=True, alpha = 0.5)
    ax.set_title('size = %d' % sample_size)
    
    sample_size = sample_size * 2
    
plt.show()
```

```python
mean = [0,0]
cov = [[1,0.75],[0.75,1]]
x, y = numpy.random.multivariate_normal(mean, cov, 5000).T
```

```python
# use gridspec to partition the figure into subplots
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

lower_right.scatter(x, y, s = 1)
top_histogram.hist(x, bins=100, normed = True)
lower_right.axis('off')
side_histogram.hist(y, bins=100, orientation='horizontal', normed = True)
plt.show()
```

```python


import pandas as pd
import folium
from matplotlib.colors import Normalize, rgb2hex
import matplotlib.cm as cm

```

```python
data = pd.read_csv('http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.csv')
norm = Normalize(data['mag'].min(), data['mag'].max())

fmap = folium.map.LegacyMap(location=[48, -102], zoom_start=3)
for eq in data.iterrows():
    color = rgb2hex(cm.OrRd(norm(float(eq[1]['mag']))))
    folium.CircleMarker([eq[1]['latitude'], eq[1]['longitude']], 
                    popup=eq[1]['place'], 
                    radius=2*float(eq[1]['mag']),
                    fill_color=color).add_to(fmap)
fmap.save('results/earthquake.html')
# need to replace CDN with https URLs
with open('results/earthquake.html', 'r') as f:
    contents = f.read()
    contents = contents.replace("http://cdn.leafletjs.com/leaflet-0.5/", "//cdnjs.cloudflare.com/ajax/libs/leaflet/0.7.7/")
    with open('results/earthquake2.html', 'w') as f:
        f.writelines(contents)
```
