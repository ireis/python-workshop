### Views Vs. copies
* Note that in the above commands b is not a new array, but a 'view' of a, i.e changing a will change b and vice versa
* http://scipy-cookbook.readthedocs.io/items/ViewsVsCopies.html

```python
print('View:')
a = numpy.arange(-10,10)
b = a[-10:]
print(b)
a[-10:] = 0
print(b, '\n')
```

To create a copy use numpy.copy():

```python
print('Copy:')
a = numpy.arange(-10,10)
b = a[-10:].copy()
print(b)
a[-10:] = 0
print(b, '\n')
```

An exception is b = a[[1,3,5,7]], which creates a copy by default  (it's considered 'fancy indexing', or not a slice, see link above).

```python
print('fancy indexing:')
a = numpy.arange(-10,10)
b = a[[1,3,5,7]]
print(b)
a[[1,3,5,7]] = 0
print(b)
```

### Histograms

Draw numbers from a normal distributions, loc = mean, scale = sigma:

```python
n1 = numpy.random.normal(loc = 0, scale=  1, size=10000)
n2 = numpy.random.normal(loc = 0, scale=  2, size=10000)
```

```python
plt.figure()
plt.hist(n1, bins = 100, normed=True, alpha = 0.5, color = 'blue',   label = r'$\sigma = 1$')
plt.hist(n2, bins = 100, normed=True, alpha = 0.5, color = 'purple', label = r'$\sigma = 2$')
plt.legend()
plt.show()
```

### Sub-plots
Example for a grid of plots in one figure, the size of the grid is controlled by nrows, ncols. sharex - same x axis for all subplots

```python
f, axes = plt.subplots(ncols=3, nrows=3, figsize = (10,7), sharex=True)

sample_size = 100
for ax in axes.flatten():
    n = numpy.random.normal(loc = 0, scale=  1, size=sample_size)
    
    ax.hist(n, bins = 100, normed=True, alpha = 0.5)
    ax.set_title('size = %d' % sample_size)
    
    sample_size = sample_size * 2
    
plt.show()
```

#### Another subplot example - marginal distributions

```python
mean = [0,0]
cov = [[1,0.75],[0.75,1]]
x, y = numpy.random.multivariate_normal(mean, cov, 5000).T
```

```python
# use gridspec to partition the figure into subplots
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

lower_right.scatter(x, y, s = 1)
top_histogram.hist(x, bins=100, normed = True)
lower_right.axis('off')
side_histogram.hist(y, bins=100, orientation='horizontal', normed = True )
plt.show()
```

### Examples with real data

```python
import pandas as pd
from astropy.table import Table
from astropy.io import fits 

fname = 'data/DR12Q_BAL.fits'
#https://data.sdss.org/sas/dr13/env/BOSS_QSO/DR12Q/DR12Q_BAL.fits
hdulist = fits.open(fname)

tbdata = hdulist[1].data
tbdata = Table(tbdata)
hdulist.close()
```

### Density map

```python
from matplotlib.colors import LogNorm
import matplotlib.cm as cm

x = tbdata['Z_VI']
y = tbdata['Z_PIPE']

xmn = x.min()
xmx = x.max()
ymn = y.min()
ymx = y.max()

xlim = [xmn,xmx]
ylim = [ymn,ymx]

# Calculate the 2d density histogram
[hist_tmp, xbin, ybin] = numpy.histogram2d(x, y, bins=[100,100], range=[xlim,ylim], normed=False, weights=None)

plt.figure(figsize=(10,7))

# Plot the histogram
plt.imshow(hist_tmp.T[:][::-1] + 0.01, extent=(numpy.amin(xlim), numpy.amax(xlim), numpy.amin(ylim), numpy.amax(ylim)), 
           cmap=cm.viridis, norm=LogNorm(), aspect='auto')

# Edit colorbar tick etc.
cb = plt.colorbar()
cb.ax.tick_params(labelsize=18)
cb.set_ticks([1,10,100,1000])

plt.xlabel('Z from visual inspection', fontsize = 20)
plt.ylabel('Z from SDSS pipeline', fontsize = 20)
plt.xlim(xlim)
plt.ylim(ylim)

cb.ax.set_ylabel('Number of quasars', rotation=90, fontsize = 20)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.show()
```
