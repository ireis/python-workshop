## Astropy

The astropy package (alternatively known as the "core" package) contains various classes, utilities, and a packaging framework intended to provide commonly-used astronomy tools. It is divided into a variety of sub-packages, which are described in detail in the documentation: http://docs.astropy.org/en/stable/. There are various other packages that are affiliated with the Astropy development community: http://www.astropy.org/affiliated/.

The core also provides this documentation, and a variety of utilities that simplify starting other python astronomy/astrophysics packages. The main utilities in astropy include:

### Core data structures and transformations

* Constants (astropy.constants)
* Units and Quantities (astropy.units)
* N-dimensional datasets (astropy.nddata)
* Data Tables (astropy.table)
* Time and Dates (astropy.time)
* Astronomical Coordinate Systems (astropy.coordinates)
* World Coordinate System (astropy.wcs)
* Models and Fitting (astropy.modeling)
* Analytic Functions (astropy.analytic_functions)
* Connecting up: Files and I/O

### Unified file read/write interface

* FITS File handling (astropy.io.fits)
* ASCII Tables (astropy.io.ascii)
* VOTable XML handling (astropy.io.votable)
* Miscellaneous: HDF5, YAML, pickle (astropy.io.misc)
* Virtual Observatory Access (astropy.vo)

### Astronomy computations and utilities

* Cosmological Calculations (astropy.cosmology)
* Convolution and filtering (astropy.convolution)
* Data Visualization (astropy.visualization)
* Astrostatistics Tools (astropy.stats)

## Example 1 - query SDSS databases and Simbad
We will start with an example where we have a PLATE-MJD-FIBER identifier of an object from the SDSS. We want to download and show the objects spectrum, then extract its coordinates, and finally look for references for this object in SIMBAD.

```python
from astropy.coordinates import SkyCoord
from astroquery.vizier import Vizier
from astropy import units as u
from astroquery.simbad import Simbad
from astroquery.sdss import SDSS

plate = 2348
mjd = 53757
fiber = 500

# query SDSS database for this object
spectra = SDSS.get_spectra(plate=plate, mjd=mjd, fiberID=fiber)
print("got the spectra, length: %s" % len(spectra))

# plot the spectrum
data = spectra[0][1].data
wavelength = 10 ** data['loglam']
flux = data['flux']

plt.rcParams['figure.figsize'] = 10, 4
plt.step(wavelength, flux, "k")
plt.title("The Hand of Pooh")
plt.xlabel("observed wavelength")
plt.ylabel("$\\mathrm{flux\\,[10^{-17}\\, erg \\, cm^{-2} \\, s^{-1} \\, \\AA^{-1}] }$")
plt.tight_layout()
```

```python
# now I want to query the SDSS database to find the coordinates of the object
result = SDSS.query_specobj(plate=plate, mjd=mjd, fiberID=fiber, 
                            fields=['ra', 'dec'])

# define the coordinates of the object
RA_object = result['ra'][0]
DEC_object = result['dec'][0]
coords = SkyCoord(RA_object, DEC_object, unit=(u.deg, u.deg), frame='icrs')

# print the result
result
```

```python
# extract information from Simbad - does this object has a reference?
result_table = Simbad.query_region(coords, radius='0d0m20s')
bibcode = result_table['COO_BIBCODE'][0]
result_table
```

```python
# what is this bibbode?
result_table = Simbad.query_bibcode(bibcode)
# result_table = Simbad.query_bibcode(bibcode.decode("utf-8â€))
print(result_table)
```

## Example 2 - plot both redshift and age of the universe in the same figure
Each redshift corresponds to an age of the universe, so if you're plotting some quantity against redshift, it's often useful show the universe age too. The relationship between the two changes depending the type of cosmology you assume, which is where *astropy.cosmology* comes in. In this tutorial we'll show how to use the tools in *astropy.cosmology* to make a plot like this

```python
from astropy.cosmology import FlatLambdaCDM
from astropy.cosmology import Planck13

# define the cosmological parameters we want to work with
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)

# define redshifts, and calculate angular diameter distance
redshift = numpy.arange(0, 6, 0.1)
angular_diameter_distance = cosmo.angular_diameter_distance(redshift)
planck_angular_diameter_distance = Planck13.angular_diameter_distance(redshift)

# what is the units of the angular diameter distance?
print("The units of the angular diameter distance: %s" % angular_diameter_distance.unit)
```

```python
from astropy.cosmology import z_at_value

# now return the age of the universe for each redshift
uni_age = cosmo.age(redshift)
print("The units of the age: %s" % uni_age.unit)

# finaly, plot it all on a single plot!
plt.rcParams['figure.figsize'] = 10, 4

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

ax1.plot(redshift, angular_diameter_distance, "b", label="FlatCosmo")
ax1.plot(redshift, planck_angular_diameter_distance, "r", label="Planck13")
ax1.set_xlabel("redshift")
ax1.set_xticklabels(map(str, ax1.get_xticks()))
ax1.set_ylabel("Angular diameter distance (Mpc)")
ax1.legend(loc="lower right")
ax1.set_yticks(numpy.arange(0, 2000, 500))

ages = np.array([13, 10, 8, 6, 5, 4, 3, 2, 1.5, 1.2, 1])*u.Gyr
ageticks = [z_at_value(cosmo.age, age) for age in ages]
ax2.set_xticks(ageticks)
ax2.set_xticklabels(['{:g}'.format(age) for age in ages.value])
ax2.set_xlabel("Age of the universe (Gyr)")
```

## Example 3 - Galaxy mass measurement from velocity dispesrion
In this first example, we will use *Quantity* objects to estimate a hypothetical galaxy's mass, given its half-light radius and radial velocities of stars in the galaxy.
Lets assume that we measured the half light radius of the galaxy to be 29 pc projected on the sky at the distance of the galaxy. This radius is often called the "effective radius", so we will store it as a Quantity object with the name Reff. The easiest way to create a Quantity object is just by multiplying the value with its unit. Units are accessed as u."unit", in this case u.pc.

```python
import astropy.units as u
from astropy.constants import G, h, k_B
```

```python
Reff = 29 * u.pc
# Reff = u.Quantity(29, unit=u.pc)

print("""Half light radius
value: {0}
unit: {1}""".format(Reff.value, Reff.unit))
```

Furthermore, we can convert the radius in parsecs to any other unit of length using the to() method. Here, we convert it to meters.

```python
print("{0:.3g}".format(Reff.to(u.kpc)))
# print("{0:.3g}".format(Reff.to(u.m)))
# print("{0:.3g}".format(Reff.to(u.cm)))
# print("{0:.3g}".format(Reff.to(u.mpc)))
```

Next, we will first create a synthetic dataset of radial velocity measurements, assuming a normal distribution with a mean velocity of 206 km/s and a velocity dispersion of 4.3 km/s.

```python
vmean = 206
sigin = 4.3
v = numpy.random.normal(vmean, sigin, 500)*u.km/u.s
```

```python
print("""First 10 radial velocity measurements: 
{0}
{1}""".format(v[:10], v.to(u.m/u.s)[:10]))
```

```python
plt.figure()
plt.hist(v, bins=20, histtype="step")
plt.xlabel("Velocity (km/s)")
plt.ylabel("N")
```

Next, we calculate the velocity dispersion of the galaxy. This demonstrates how you can perform basic operations like subtraction and division with Quantity objects, and also use them in standard numpy functions such as mean() and size(). They retain their units through these operations just as you would expect them to.

```python
sigma = numpy.sqrt(numpy.sum((v - numpy.mean(v))**2) / numpy.size(v))
print("Velocity dispersion: {0:.2f}".format(sigma))
```

Now, lets estimate the galaxy mass from the effective radius and the velocity dispersion.
We will use the following estimate for the mass: 

$M_{1/2}=4\sigma^{2}R_{eff}/G$

```python
M = 4*sigma**2*Reff/G
M
```

The result is in a composite unit, so it's not really obvious it's a mass. However, it can be decomposed to cancel all of the length units using the decompose() method.

```python
M.decompose()
```

We can also easily express the mass in whatever form you like - solar masses are common in astronomy, or maybe you want the default SI and CGS units.

```python
print("""Galaxy mass
in solar units: {0:.3g}
SI units: {1:.3g}
CGS units: {2:.3g}""".format(M.to(u.Msun), M.si, M.cgs))
```

Or, if you want the log of the mass, you can just use numpy.log10 as long as the logarithm's argument is dimensionless.

```python
print(numpy.log10(M / u.Msun))
```

Similarly, we can use *Quantity* and Kepler's law in the form given below to determine the (circular) orbital speed of the Earth around the sun in km/s. 

$ v = \sqrt{\frac{G M_{\odot}}{r}}$

```python
import astropy.constants
velocity = numpy.sqrt(G*astropy.constants.M_sun / u.au)
velocity
```
